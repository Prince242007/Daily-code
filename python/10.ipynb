```python
import matplotlib.pyplot as plt 
import numpy as np 
x = np.linspace(0.1, 2 * np.pi, 41) 
y = np.exp(np.sin(x)) 
plt.title("Stem Chart")
plt.stem(x, y) 
plt.show() 
```

```python
import matplotlib.pyplot as plt 
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 
# This will plot a simple bar chart
plt.bar(x, y)
# Title to the plot
plt.title("Bar Chart")
# Adding the legends
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt 
x = [1, 2, 3, 4, 5, 6, 7, 4] 
# This will plot a simple histogram
plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7])
# Title to the plot
plt.title("Histogram")
# Adding the legends
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt 
# data to display on plots 
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.scatter(x, y)
plt.legend("A")
plt.title("Scatter chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```

```python
import matplotlib.pyplot as plt 
# data to display on plots 
x = [1, 2, 3, 4] 
# this will explode the 1st wedge
# i.e. will separate the 1st wedge
# from the chart
e  =(0.1, 0, 0, 0)
# This will plot a simple pie chart
plt.pie(x, explode = e)
# Title to the plot
plt.title("Pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting the line
plt.plot(x, y)

# Adding labels and title
plt.xlabel("X Axis - Numbers")
plt.ylabel("Y Axis - Doubles")
plt.title("Simple Line Chart")

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data for first line
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
# Data for second line
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
plt.plot(x1, y1, label="y = 2x")       # First line with legend
plt.plot(x2, y2, label="y = x^2")      # Second line with legend
# Adding labels and title
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Two Lines on Same Plot")
# Adding legend
plt.legend()
# Show the chart
plt.show()
```
