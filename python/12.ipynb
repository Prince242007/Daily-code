```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
# Parameters
fs = 500  # Sampling frequency
t = np.linspace(-1, 1, fs, endpoint=False)  # Time array
# 1. Unit Step Signal
unit_step = np.heaviside(t, 1)
plt.figure(figsize=(12, 12))
plt.subplot(4, 2, 1)
plt.plot(t, unit_step)
plt.title('Unit Step Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Parameters
n = np.arange(0, 20)  # Discrete time array (0 to 19)
signal = np.sin(0.2 * np.pi * n)  # Example discrete-time signal (sine wave)
# Delay the signal by 3 samples
delay = 3
delayed_signal = np.zeros_like(signal)
delayed_signal[delay:] = signal[:-delay]
# Advance the signal by 3 samples
advance = 3
advanced_signal = np.zeros_like(signal)
advanced_signal[:-advance] = signal[advance:]
# Plot the original and shifted signals
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 2)
plt.stem(n, delayed_signal)
plt.title(f'Delayed Signal (by {delay} samples)')
plt.xlabel('n (Discrete Time)')
plt.ylabel('Amplitude')
# Advanced Signal
plt.subplot(3, 1, 3)
plt.stem(n, delayed_signal)
plt.title(f'Advanced Signal (by {advance} samples)')
plt.xlabel('n (Discrete Time)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 20  # Sampling frequency for discrete-time signal
f = 20  # Sampling frequency for discrete-time signal
t_continuous = np.linspace(0, 1, 1000)  # Time array for continuous signals
t_discrete = np.arange(0, 1, 1/fs)  # Discrete time array
discrete_time_signal = np.sin(2 * np.pi * f * t_discrete)
plt.figure(figsize=(12, 10))
plt.subplot(4, 1, 1)
plt.subplot(4, 1, 2)
plt.stem(t_discrete, discrete_time_signal)
plt.title('Discrete-Time Signal (Sampled Sine Wave)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# sampling details
fs = 1000          # sampling frequency
t = np.arange(0, 1, 1/fs)   # time from 0 to 1 sec

# two sine waves
s1 = np.sin(2*np.pi*5*t)    # 5 Hz sine wave
s2 = np.sin(2*np.pi*10*t)   # 10 Hz sine wave

# add the signals
s_sum = s1 + s2

# plot
plt.plot(t, s_sum)
plt.title("Sum of 5 Hz and 10 Hz Sine Waves")
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# sampling info
fs = 500
t = np.arange(0, 2, 1/fs)

# signals
x1 = np.sin(2*np.pi*5*t)      # 5 Hz sine
x2 = np.cos(2*np.pi*10*t)     # 10 Hz cosine

# element wise multiply
x_mul = x1 * x2

# plot
plt.plot(t, x_mul)
plt.title("Product of 5 Hz Sine and 10 Hz Cosine")
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# sampling frequency and time
fs = 1000
t = np.arange(0, 1, 1/fs)

# original 5 Hz sine wave
x = np.sin(2*np.pi*5*t)

# time shift of 0.1 seconds
shift = int(0.1 * fs)     # convert seconds into number of samples

# create shifted signal
x_shift = np.zeros_like(x)
x_shift[shift:] = x[:-shift]

# plot both signals
plt.plot(t, x, label="Original")
plt.plot(t, x_shift, label="Shifted by 0.1 sec")
plt.title("Original vs Time-Shifted 5 Hz Sine Wave")
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# sampling frequency and time axis
fs = 1000
t = np.arange(0, 1, 1/fs)

# original 10 Hz sine wave
x = np.sin(2*np.pi*10*t)

# scale amplitude by factor 3
x_scaled = 3 * x

# plot both signals
plt.plot(t, x, label="Original 10 Hz Sine")
plt.plot(t, x_scaled, label="Scaled (Amplitude  3)")
plt.title("Original vs Scaled 10 Hz Sine Wave")
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# sampling frequency and time
fs = 1000
t = np.arange(0, 1, 1/fs)

# original 5 Hz sine wave
x = np.sin(2*np.pi*5*t)

# reverse the signal in time
x_rev = x[::-1]

# plot both signals
plt.plot(t, x, label="Original Signal")
plt.plot(t, x_rev, label="Reversed Signal")
plt.title("Original vs Time-Reversed 5 Hz Sine Wave")
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
